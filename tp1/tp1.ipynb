### **Parte 2**


- Librerias a utilizar y carga de imágenes en escala de grises


```python
%matplotlib inline
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img1 = cv.imread('assets/TP1/img1_tp.png', cv.IMREAD_GRAYSCALE)
img2 = cv.imread('assets/TP1/img2_tp.png', cv.IMREAD_GRAYSCALE)
```

- Defino funciones para mostrar la imagen y el histograma


```python
def showImage(img):
    plt.figure()
    plt.imshow(img,cmap='gray')
    plt.show()

def showHistrogram(img, binsAmount):
    hist_np, bins = np.histogram(img,binsAmount, range=(0, 256))
    plt.figure()
    plt.title('Histograma con NumPy')
    plt.xlabel('Nivel de gris')
    plt.ylabel('Frecuencia')
    plt.plot(hist_np)
    plt.grid(True)
    plt.show()
```

**Muestro Imagen 1**


```python
showImage(img1)
```

**Histograma Imagen 1**


```python
showHistrogram(img1,256)
```

**Muestro Imagen 2**


```python
showImage(img2)
```

**Histograma Imagen 1**


```python
showHistrogram(img2,256)
```
